3. IMPLEMENTAÇÃO
− Código completo da classe FiniteAutomata
− Funções auxiliares


```python

class FiniteAutomata :
    #delta : Dict [( str , str ) , ( str, str, str )] = None
    estados = []
    alfabeto = []
    func_transicao = {}
    estado_inicial = ""
    estado_final = []

# Inicialização
    def __init__ (estados, alfabeto, func_transicao, estado_inicial, estado_final ):
      #Recebe dados do AF
      print("Informe o conjunto de estados: ", end = "")
      estados = input().split( )

      print("Informe o alfabeto de entrada: ", end = "")
      alfabeto = input().split( )

      print("Informe o estado inicial: ", end = "")
      estado_inicial = input().split( )

      print("Informe o conjunto de estados de finais: ", end = "")  #estados de 'aceitação'
      estado_final = input().split( )

#Executa um passo do AF
    def step( alfabeto, estados ):
      print("Defina as funções de transição: ", end = "")
      for estado in estados:
        for simbolo in alfabeto:
          print(f"\t{simbolo}")
          print(f"{estado}\t--->\t", end="")
          proximo_estado = input()  #no {estado} ---> proximo_estado

          if proximo_estado == ".":
            func_transicao[(estado, simbolo)] = None
          else:
            func_transicao[(estado, simbolo)] = proximo_estado
            #Retorna : keep_running


#Executa o AF para a fita fornecida
    def execute( ):
        print("Informe uma linguagem/fita a ser reconhecida: ", end="")
        entrada = input()

        estado_atual = estado_inicial

        for simbolo in entrada:
          print(f"Estado atual: {estado_atual}")
          print(f"Entrada atual: {simbolo}")
          print(f"Proximo estado: {func_transicao[(estado_atual, simbolo)]}")

          estado_atual = func_transicao[(estado_atual, simbolo)]

          if estado_atual == None:
            print("O automato não  reconhece a linguagem")
            break

          if estado_atual in estado_final:
            print("Reconhece") ##Retorna : ( accept )
          else:
            print("Não reconhece")


    __init__(estados, alfabeto, func_transicao, estado_inicial, estado_final)
    step(alfabeto, estados)
```
